# **Classification And Regression**

```python
dr_lj14= df['lanejumping2014']
dr_lj14=np.array(dr_lj14)
```

```python
dr_lj16= df['lanejumping2016']
dr_lj16=np.array(dr_lj16)
```

```python
Xy = df
X = dr_lj14.reshape(-1,1)
y = dr_lj16.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
result=reg.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,result, c='black')
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = (0.8))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_predict = knn.predict(X_test)
y_predict
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_train.dtype
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('lanejumping2014')
plt.ylabel('lanejumping2016')
s = [0,5000] #reference line to show how these two amounts (test and preditced) are different
plt.plot(s,s, color='red')
plt.show()
```

```python
c = [i for i in range(1,len(y_test)+1,1)]
plt.plot(c,y_test-y_pred,color='green',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
```

```python
import sklearn
from sklearn import metrics
mse = metrics.mean_squared_error(y_test, y_pred)
mse
```

```python
sklearn.metrics.r2_score(y_test, y_pred)
```
